```python
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq
import torch
import librosa

# Load the processor and model
processor = AutoProcessor.from_pretrained("openai/whisper-small")
model = AutoModelForSpeechSeq2Seq.from_pretrained("openai/whisper-small")

def transcribe_with_huggingface_whisper_to_file(audio_path, output_file):
    # Load audio file (16kHz, mono)
    audio, sr = librosa.load(audio_path, sr=16000)

    # Prepare inputs, ensuring English transcription (language='en')
    inputs = processor(audio, return_tensors="pt", sampling_rate=16000, language='en')

    # Generate transcription without using attention_mask
    with torch.no_grad():
        generated_ids = model.generate(inputs["input_features"])

    # Decode the output into text
    transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].lower()

    # Save the transcription to file2.txt
    with open(output_file, "w") as f:
        f.write(transcription)
```

```python
def transcribe_with_custom_model(input_file, output_file):
    r = sr.Recognizer()
    try:
        with sr.AudioFile(input_file) as source:
            r.adjust_for_ambient_noise(source, duration=0.2)
            audio_length = source.DURATION  # Use total duration of audio

            chunk_duration = min(30, audio_length)  # Choose smaller chunks based on total duration
            num_chunks = int(audio_length // chunk_duration) + 1

            with open(output_file, "w") as f:
                for i in range(num_chunks):
                    start_time = i * chunk_duration
                    audio_chunk = r.record(source, duration=chunk_duration)

                    try:
                        text = r.recognize_google(audio_chunk).lower()
                        # print(f"Chunk {i+1}/{num_chunks}: {text}")
                        f.write(f"{text} ")
                    except sr.UnknownValueError:
                        # print(f"Unrecognized speech in chunk {i+1}")
                        f.write("[Unrecognized speech]\n")
                    except sr.RequestError as e:
                        # print(f"Error processing chunk {i+1}: {e}")
                        f.write("[Processing error]\n")

    except FileNotFoundError:
        print("Audio file not found.")
    except Exception as e:
        print(f"Error transcribing audio file: {e}")
```

```python
from jiwer import wer, cer

# Calculate WER and CER for custom model
custom_wer = wer(ground_truth_text, custom_transcription)
custom_cer = cer(ground_truth_text, custom_transcription)

# Calculate WER and CER for Hugging Face Whisper model
whisper_wer = wer(ground_truth_text, whisper_transcription)
whisper_cer = cer(ground_truth_text, whisper_transcription)

# Print WER and CER for both models
print(f"Custom Model WER: {custom_wer:.2f}, CER: {custom_cer:.2f}")
print(f"Whisper Model WER: {whisper_wer:.2f}, CER: {whisper_cer:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for bar chart
metrics = ['WER', 'CER']
custom_metrics = [custom_wer, custom_cer]
whisper_metrics = [whisper_wer, whisper_cer]

# Create bar chart for WER and CER comparison
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(metrics))

# Plot custom model metrics
plt.bar(index, custom_metrics, bar_width, label='Custom Model', color='b')

# Plot Whisper model metrics
plt.bar([i + bar_width for i in index], whisper_metrics, bar_width, label='Whisper Model', color='g')

plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.title('WER and CER Comparison')
plt.xticks([i + bar_width / 2 for i in index], metrics)
plt.legend()

# Show plot
plt.show()
```
